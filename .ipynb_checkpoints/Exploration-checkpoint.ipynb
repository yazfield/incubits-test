## BOW and TF-IDF
First I constructed bag of words (BOW) and TF-IDF vectors. 
I tried to reduce the number of words by using stemming, excude stop words, limit vocabulary size due to the number of examples in the dataset and to speedup training too.

```python
RANDOM_STATE = 7878
VECTORS_PATH = path.join(DATA_PATH, 'vectors')
```

```python
# used a stemmer to reduce tokens
stemmer = SnowballStemmer("english", ignore_stopwords=True)
class StemmedCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(StemmedCountVectorizer, self).build_analyzer()
        return lambda doc: ([stemmer.stem(w) for w in analyzer(doc)])
```

```python
# Training target
y = training_data['Class'].values
```

### BOW vectors

```python
TRAIN_BOW_90K = path.join(VECTORS_PATH, 'train_bow_90k.npz')
vectorizer = StemmedCountVectorizer(tokenizer=nltk.word_tokenize, stop_words='english', 
                             max_df=0.8, max_features=90000, binary=False, min_df=0.03)
if path.exists(TRAIN_BOW_90K):
    X_bow = load_npz(TRAIN_BOW_90K)
else:
    X_bow = vectorizer.fit_transform(training_data['Text'].values)
    save_npz(TRAIN_BOW_90K, X_bow)
    
print('Training BOW shape:', X_bow.shape)
print('Targets shape:', y.shape)
```

### TF-IDF vectors

```python
tfidf_transformer = TfidfTransformer()
X_tfidf = tfidf_transformer.fit_transform(X_bow)
```

### Bigram BOW vectors

```python
TRAIN_BIGRAM_BOW_90K = path.join(VECTORS_PATH, 'train_bigram_bow_90k.npz')
bigram_vectorizer = StemmedCountVectorizer(tokenizer=nltk.word_tokenize, stop_words='english', ngram_range=(1, 2),
                             max_df=0.8, max_features=90000, binary=False, min_df=0.03)
if path.exists(TRAIN_BIGRAM_BOW_90K):
    X_bigram_bow = load_npz(TRAIN_BIGRAM_BOW_90K)
else:
    X_bigram_bow = bigram_vectorizer.fit_transform(training_data['Text'].values)
    save_npz(TRAIN_BIGRAM_BOW_90K, X_bigram_bow)
    
print('Training bigram BOW shape:', X_bigram_bow.shape)
print('Targets shape:', y.shape)
```

### Bigram TF-IDF vectors

```python
bigram_tfidf_transformer = TfidfTransformer()
X_bigram_tfidf = bigram_tfidf_transformer.fit_transform(X_bigram_bow)
```

## Doc2Vev
I trained some Doc2Vec models of different sizes and settings (with / without stopwords) using the script **doc2vec.py**.

```python
STOPWORDS = False
def clean(text):
    # TODO: optimize this
    n = []
    stopw = stopwords.words('english')
    for txt in text.strip().lower().split():
        txt = re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', txt)
        txt = re.sub('[\W_]', '', txt)
        txt = re.sub(' [0-9]+ ', ' ', txt)
        txt = ' '.join(txt.split())
        txt = re.sub('^[0-9]+ ', ' ', txt)
        txt = ' '.join(txt.split())
        if re.match('^[0-9]+$', txt):
            continue
        if STOPWORDS and txt in stopw:
            continue
        if txt:
            n.append(txt)
    return ' '.join(n)
```

Run this script to generate the vectors:
`python3 vectorize.py  --vectorizer doc2vec --size 1000 --filename 'data/vectors/doc2vec500' --stop_words None`

# Training

```python
def show_prediction_result(classifier, X, y):
    probas = classifier.predict_proba(X)
    indices = np.argmax(probas, axis=1)
    classes = np.unique(y)
    predictions = classes[indices]
    print('Log loss: {}'.format(log_loss(y, probas)))
    print(classification_report(y, predictions))
    plot_confusion_matrix(y, predictions)
    
def evaluate(classifier, X, y, X_test=None, y_test=None):
    classifier.fit(X, y)
    show_prediction_result(classifier, X, y)
    if X_test is not None:
        print('')
        print('Test set:')
        show_prediction_result(classifier, X_test, y_test)
    return classifier
        
```

```python
test_variants = pd.read_csv(TEST_VARIANTS_PATH)
test_text = pd.read_csv(TEST_TEXT_PATH, sep="\|\|", engine="python", skiprows=1, names=["ID", "Text"])
test_y_data = pd.read_csv('data/stage1_solution_filtered.csv')
test_data = pd.merge(test_text, test_variants, on='ID')
test = test_data.loc[test_data['ID'].isin(test_y_data['ID'].values)]
yt = np.argmax(test_y_data.values[:, 1:], axis=1) + 1
```

### MLP and LSTM
After some hyperparameter search, the following models performed best.

- MLP with Bigram TF-IDF
- LSTM with Bigram TF-IDF

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping
batch_size = 128
num_classes = 9
epochs = 50

x_train = X_bigram_tfidf.todense()
x_test = Xtest_bigram_tfidf.todense()
y_train = keras.utils.to_categorical(y - 1, num_classes)
y_test = keras.utils.to_categorical(yt - 1, num_classes)

model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(24539,)))
model.add(Dropout(0.2))
model.add(Dense(512, activation='tanh'))
model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_test, y_test),
                   callbacks=[EarlyStopping(patience=8)])
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
from keras.layers import LSTM

batch_size = 64
num_classes = 9
epochs = 50
hidden = 128

x_train = np.array(X_bigram_tfidf.todense()).reshape((-1, 1, 24539))
x_test = np.array(Xtest_bigram_tfidf.todense()).reshape((-1, 1, 24539))
y_train = keras.utils.to_categorical(y - 1, num_classes)
y_test = keras.utils.to_categorical(yt - 1, num_classes)

model = Sequential()

model.add(LSTM(hidden, dropout=0.3, recurrent_dropout=0.3, input_shape=(1, 24539)))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_test, y_test),
                    callbacks=[EarlyStopping(patience=8)])
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
